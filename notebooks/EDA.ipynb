```python
#Load all environments
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import streamlit as st
# Read the datasetâ€™s CSV file into a Pandas DataFrame
cars_df = pd.read_csv('vehicles_us.csv')
```

```python
# Display basic information about the dataset
cars_df.info()
```

```python
# Display summary statistics
cars_df.describe()
```

```python
# Check for missing values
cars_df.isnull().sum()
```

```python
# Remove missing values 
cars_df.dropna(inplace=True)
# Check for missing values agin
cars_df.isnull().sum()
```

```python
# Plot histogram of prices
plt.hist(cars_df['price'], bins=50)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
# Convert model_year to integer cars_df['model_year'] = cars_df['model_year'].astype(int) # Scatter Plot of Price by Car Type plt.figure(figsize=(12, 6)) sns.scatterplot(x='type', y='price', data=cars_df) plt.xlabel('Car Type') plt.ylabel('Price') plt.title('Price by Car Type') plt.xticks(rotation=45) plt.show()# Plot scatter plot of price vs. model year
sns.scatterplot(x='model_year', y='price', data=cars_df)
plt.xlabel('Model Year')
plt.ylabel('Price')
plt.title('Price vs. Model Year')
plt.show()
```
